```python
def process_piomass(year):
    
    binary_dir = f'binaries/heff.H{year}'
    
    with open(binary_dir, mode='rb') as file:
    
        fileContent = file.read()
        data = struct.unpack("f" * (len(fileContent)// 4), fileContent)
        
        native_data = np.full((12,360,120),np.nan)
        EASE_data = np.full((12,361,361),np.nan)
        

        for month in range(1,13):
            
            start = (month-1)*(360*120)
            end = month*(360*120)
            thickness_list = np.array(data[start:end])
            
            gridded = thickness_list.reshape(360,120)
            native_data[month-1,:,:] = gridded
            
#             cartoplot(grid['lon'],grid['lat'],gridded)

            regridded = regrid(gridded,grid['lon'],grid['lat'],mask.get('lon'),mask.get('lat'))
            EASE_data[month-1,:,:] = regridded
            
#             cartoplot(mask.get('lon'),mask.get('lat'),regridded)
        
        ############################################################
        
        # OUTPUT IN NATIVE RESOLUTION
        
        ds_native = xr.Dataset( data_vars={'thickness':(['t','x','y'],native_data)},

                         coords =  {'lon':(['x','y'],grid['lon']),
                                    'lat':(['x','y'],grid['lat']),
                                    'month':(['t'],np.array(range(1,13)))})

        output_dir = f'/home/robbie/Dropbox/SM_Thickness/data/piomass/360x120/'

        ds_native.to_netcdf(f'{output_dir}{year}.nc','w')
        
        ############################################################

        ds_regridded = xr.Dataset( data_vars={'thickness':(['t','x','y'],EASE_data)},

                         coords =  {'lon':(['x','y'],mask.get('lon')),
                                    'lat':(['x','y'],mask.get('lat')),
                                    'month':(['t'],np.array(range(1,13)))})

        output_dir = f'/home/robbie/Dropbox/SM_Thickness/data/piomass/361x361/'

        ds_regridded.to_netcdf(f'{output_dir}{year}.nc','w')
        
for year in trange(2002,2020):
    process_piomass(year)
```

```python
def regrid_piomass_to_EASE(year,month):
    
        data_dir = '/home/robbie/Dropbox/SM_Thickness/data/piomass/361x361/'

        with xr.open_dataset(f'{data_dir}{year}.nc') as data:

            ds_month = data.where(int(month) == data.month, drop=True)

            return(ds_month)
    

    
field = regrid_piomass_to_EASE(2005,2)


cartoplot(field['lon'],field['lat'],field['thickness'][0])
```
